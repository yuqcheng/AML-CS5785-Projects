```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dt = np.genfromtxt('hw3_2_faithuful.txt', dtype=[int,float,int])
```

```python
col_index = []
eruptions = []
writting = []
for i in list(dt):
    col_index.append(i[0])
    eruptions.append(i[1])
    writting.append(i[2])
```

```python
data = pd.DataFrame({'index': col_index, 'eruptions':eruptions,'writing':writting})
```

```python
data = data.set_index(data.columns[0])
```

```python
two_dim_vector = data.values
```

```python
plt.scatter(two_dim_vector[:,0], two_dim_vector[:,1])
plt.xlabel("Eruptions")
plt.ylabel("Writing")
plt.show()
```
